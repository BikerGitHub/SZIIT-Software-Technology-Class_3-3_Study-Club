```python
import numpy as np
import matplotlib.pyplot as plot
csv_array=np.loadtxt('mall_sales.csv',dtype=str,delimiter=',')
head=csv_array[0]
csv_array=csv_array[1:,1:]
csv_array=csv_array.astype(float)
```

```python
print(csv_array)
```

```python
sales_r=lambda x:np.diff(csv_array[:,x])/csv_array[:-1,x]
covm=np.cov(sales_r(0),sales_r(3))
print(covm,covm.diagonal(),covm.trace(),sep='\n')
```

```python
r=covm/(sales_r(0).std(ddof=1)*sales_r(3).std(ddof=1))
# np.std([1,2,3],ddof=1)
# covm/(sales_r(0).std()*sales_r(3).std())
print(r)
```

```python
t=np.arange(len(csv_array)-1)
plot.plot(t,sales_r(0),lw=1)
plot.plot(t,sales_r(3),lw=1)
plot.show()
```
