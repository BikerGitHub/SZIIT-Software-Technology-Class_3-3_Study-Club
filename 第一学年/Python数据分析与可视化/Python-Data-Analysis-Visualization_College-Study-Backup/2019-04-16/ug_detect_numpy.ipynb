```python
import numpy as np
import matplotlib.pyplot as plot
csv_array=np.loadtxt('ug_detect.csv',dtype=str,delimiter=',')
head=csv_array[0]
csv_array=csv_array[1:]
csv_array[csv_array=='']='999'
csv_array=csv_array.astype(float)
```

```python
print(head)
print(csv_array[:9])
```

```python
for index in range(1,len(csv_array)):
    for index_n in range(len(csv_array[index])):
        if csv_array[index,index_n]>888:
            csv_array[index,index_n]=(csv_array[index-1,index_n]+csv_array[index+1,index_n])/2
```

```python
plot.plot(csv_array[1:,1])
plot.plot(csv_array[1:,1],'pr')
plot.show()
```
