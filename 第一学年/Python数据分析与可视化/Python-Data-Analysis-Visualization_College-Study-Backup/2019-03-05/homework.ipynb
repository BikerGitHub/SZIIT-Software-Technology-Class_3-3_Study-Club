### 绘制函数曲线
通常使用matplotlib.pyplot包

```python
# step 1: 引入需要使用的包（库）
import numpy as np
import matplotlib.pyplot as plt
```

```python
# step 2: 创建横坐标（明确绘制的图形样式）
x = np.arange(0,10,0.01)
# print(x)

# step 2: 创建纵坐标（明确绘制的图形样式）
y = np.sin(x)
# print(y)
z = np.sin(x ** 2)
```

```python
# step 3: 调用绘图包进行绘图并显示

# 设置图的大小
#plt.figure( figsize=(6,5) )
#puzzle 3:my answer
plt.figure( figsize=(16,9) )

# 设置图的名称
plt.title('Volt/time Waveforms')

# 设置横纵轴的名称
plt.xlabel("Time(s)")
plt.ylabel('Volt(v)')

# 设置纵轴的表示范围
plt.ylim(-1.3,1.2)

# puzzle 1: 设置横轴的表示范围为[-1,12]
#  to be filled
# my answer:
# plt.xlim(-1,12)
# puzzle 3:my answer
plt.xlim(0,10)

plt.plot(x,y,color='red',label='$y=sin(x)$')
#plt.plot(x,z)
#puzzle 3:my answer
plt.plot(x,z,color='blue',label='$y=sin(x^2)$')

plt.legend()
# puzzle 2: 如下代码绘制的曲线是什么？
# plt.plot(x,y,z)
# my answer:y=sin((x/100)^2)

# puzzle 3 ：为另一条曲线设置属性，图示等属性，并将图调整至美观

plt.show()
```
