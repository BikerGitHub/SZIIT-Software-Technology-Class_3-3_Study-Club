matplotlib.pypolot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10,0.01)
```

```python
plt.figure(figsize=(8,4))

plt.title('title')

plt.xlabel('x')
plt.ylabel('y')

# plt.xlim(-1,12)
# plt.ylim(-1.2,1.2)

line1=plt.plot(x,np.sin(x)+1,label='sinx+1')
line2=plt.plot(x,np.cos(x**2)+1,label='cos2x+1')

plt.legend(['sinx+1','cos2x+1'])
plt.show()
```
